```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df3 = df2.dropna()
```

```python
def val (valence):
    if ((valence >= 0.000) & (valence <= 0.500)):
        return 'Low'
    else:
        return 'High'
    
df3['val2'] = df3['valence'].apply(val)    
```

```python
df3
```

```python
X = df3[['Year', 'energy', 'tempo','danceability']]
y = df3['val2']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 6)
classifier.fit(X_train, y_train)
```

```python
predict = classifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
cm = confusion_matrix(y_test, predict)
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(cm, annot = True)
plt.xlabel('Predictions')
plt.ylabel('Actual values')
plt.title('Confusion Matrix for the Valence')
```

```python
def temp (tempo):
    if ((tempo >= 0.0) & (tempo <= 80.0)):
        return 'Slow'
    elif ((tempo > 80.0) & (tempo <= 110.0)):
        return 'Moderate'
    else:
        return 'Fast'
    
df3['tempi'] = df3['tempo'].apply(temp)    
```

```python
X = df3[['Year', 'energy', 'liveness', 'speechiness']]
y = df3['tempi']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)
predict = classifier.predict(X_test)
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
cm2 = confusion_matrix(y_test, predict)
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(cm2, annot = True)
plt.xlabel('Predictions')
plt.ylabel('Actual values')
plt.title('Confusion Matrix for the Tempo')
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators = 20, random_state = 0)
```

```python
from sklearn.metrics import accuracy_score
model2.fit(X_train, y_train)
predict = model2.predict(X_test)
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
print('Accuracy score:', accuracy_score(y_test, predict))
```

```python
names = ['Year', 'energy', 'liveness', 'speechiness']
features = pd.Series(model2.feature_importances_, index = names).sort_values(ascending = False)
features
```

```python
figure(figsize = (10, 6))
sns.barplot(features, features.index)
plt.title('Feature Importance for Tempo')
plt.xlabel('Importance')
plt.ylabel('Features')
```

```python
def acou (acc):
    if ((acc >= 0.000) & (acc <= 0.300)):
        return 0
    elif ((acc > 0.300) & (acc <= 0.700)):
        return 1
    else:
        return 2
    
df3['acc2'] = df3['acousticness'].apply(acou)    
```

```python
def dancey (dance):
    if ((dance >= 0.000) & (dance <= 0.300)):
        return 0
    elif ((dance > 0.300) & (dance <= 0.700)):
        return 1
    else:
        return 2
    
df3['dance2'] = df3['danceability'].apply(dancey)    
```

```python
X = df3[['Year', 'liveness', 'energy', 'valence', 'duration_ms']]
y = df3['dance2']
```

```python
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
model3 = DecisionTreeClassifier(criterion = 'entropy')
model3.fit(X_train, y_train)
predict = model3.predict(X_test)
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
print('Accuracy score:', accuracy_score(y_test, predict))
```
