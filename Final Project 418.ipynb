```python
def getURI(searcher):
    result = sp.search(searcher)
    try:
        uri = result['tracks']['items'][0]['uri']
        
    except:
        uri = None
    return uri

def getArtists(searcher):
    result = sp.search(searcher)
    try:
        obj2 = result['tracks']['items'][0]['artists']
        names = []
        for name in obj2:
            names.append(name['name'])
        obj2 = ",".join(names)      
    except:
        obj2 = ""
    return obj2

data['Link'] = data['Song ID'].apply(getURI)
data['Artist'] = data['Song ID'].apply(getArtists)
```
