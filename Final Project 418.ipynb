```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime
```

```python
import requests
from bs4 import BeautifulSoup

billboard = pd.read_csv('~/Desktop/Hot Stuff.csv')
print("Dataset One: ")
billboard
```

```python
billboard2 = billboard[billboard['Peak Position'] == 1.0]
print("Dataset Two: ")
billboard2.describe()
```

```python
billboard3 = billboard2.drop_duplicates()
billboard3
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_Billboard_Hot_100_number_ones_of_2020')
contents = url.text
soup = BeautifulSoup(contents, 'html.parser')
tables = soup.find_all("table", {"class": "wikitable"})

from IPython.core.display import display, HTML
display(HTML(tables[1].prettify()))
```

```python
lastYear = pd.read_html(str(tables))
lastYear = pd.DataFrame(lastYear[1])
lastYear

del lastYear['No.']
del lastYear['Issue date']
del lastYear['Ref.']

lastYear2 = lastYear.drop_duplicates()
lastYear2
```

```python
url2 = requests.get('https://en.wikipedia.org/wiki/List_of_Billboard_Hot_100_number_ones_of_2021')
contents2 = url2.text
soup2 = BeautifulSoup(contents2, 'html.parser')
tables2 = soup2.find_all("table", {"class": "wikitable"})

from IPython.core.display import display, HTML
display(HTML(tables2[0].prettify()))
```

```python

thisYear = pd.read_html(str(tables2))
thisYear = pd.DataFrame(thisYear[0])
thisYear

del thisYear['No.']
del thisYear['Issue date']
del thisYear['Ref.']


thisYear2 = thisYear.drop_duplicates()
thisYear2
```

```python
years = lastYear2.append(thisYear2)
years.rename(columns = {'Artist(s)': 'Performer'}, inplace = True)
years
```

```python
data = billboard3.append(years)
data
```

```python
data['Peak Position'] = 1.0
data
data['SongID'] = data['Song'] + data['Performer']
data['Song'] = data['Song'].replace("''", "")
length = len(data['Performer'].unique())
print("Number of artists: ",length)
length = len(data['Song'].unique())
print("Number of songs: ",length)

data['Performer'].value_counts().nlargest(10)
```
